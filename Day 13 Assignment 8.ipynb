#### 6. Filter for different condition - come up with insights from the dataset , basic plot , distribution plots


```python
diff = credit['Class'].value_counts()
diff
```

```python
fraud = credit['Class'].str.contains('1', na=False)
credit[fraud]
```

```python
not_fraud = credit['Class'].str.contains('0', na=False)
credit[not_fraud]     
```

```python
plt.xlabel("unknown features in bins")
plt.ylabel('credit card fraud data')
plt.legend(["features"]) 

plt.hist(credit[fraud]['V1'], facecolor='red',edgecolor= 'black',bins=5)
plt.hist(credit[fraud]['V2'], facecolor='green',bins=5, edgecolor= 'black',alpha = 0.5)
plt.hist(credit[fraud]['V3'], facecolor='yellow',bins=5, edgecolor= 'black',alpha = 0.6)
plt.hist(credit[fraud]['V4'], facecolor='blue',bins=5, edgecolor= 'black',alpha = 0.7)
plt.hist(credit[fraud]['V5'], facecolor='violet',bins=5, edgecolor= 'black',alpha = 0.3)
plt.show()

# fraud is dataset with Class '1'
```

```python
plt.xlabel("unknown features in bins")
plt.ylabel('credit card   non fraud data')
plt.legend(["features"]) 

# not_fraud is dataset with Class '0'

plt.hist(credit[not_fraud]['V1'], facecolor='red',edgecolor= 'black',bins=5)
plt.hist(credit[not_fraud]['V2'], facecolor='green',bins=5, edgecolor= 'black',alpha = 0.9)
plt.hist(credit[not_fraud]['V3'], facecolor='yellow',bins=5, edgecolor= 'black',alpha = 0.9)
plt.hist(credit[not_fraud]['V4'], facecolor='blue',bins=5, edgecolor= 'black',alpha = 0.9)
plt.hist(credit[not_fraud]['V5'], facecolor='violet',bins=5, edgecolor= 'black',alpha = 0.8)
plt.hist(credit[not_fraud]['V6'], facecolor='peru',bins=5, edgecolor= 'black',alpha = 0.7)

plt.show()
```

```python
plt.xlabel('0 = non farud,    1 = fraud')
plt.ylabel('credit   card   data    (fraud/nonfraud)')
credit['Class'].value_counts().plot(kind='bar', figsize=(10,8))
```
